```python
# This Data was pulled on 4/5/20 --- Data shows the previous 30 days 3/6/20 through 4/4/20

# Import packages --- READING FROM CSV FILE

import numpy as np
import pandas as pd

# Read the csv file

USAAnalyticsData3620to4420 = pd.read_csv('all-domains-30-days-4-5-20.csv')

# Reads the first five lines of the file

USAAnalyticsData3620to4420.head()
```

```python
# This Data was pulled on 5/17/20 --- Data shows the previous 30 days 4/17/20 through 5/16/20

# Import packages --- READING FROM CSV FILE

import numpy as np
import pandas as pd

# Read the csv file

USAAnalyticsData41720to51620 = pd.read_csv('all-domains-30-days-5-17-20.csv')

# Reads the first five lines of the file

USAAnalyticsData41720to51620.head()
```

```python
# Removes any visits1 that less than 4,600,000

USAAnalyticsData3620to4420 = USAAnalyticsData3620to4420[USAAnalyticsData3620to4420.visits1 > 4600000]

# Shows all of the records in the dataframe

USAAnalyticsData3620to4420.head(55)
```

```python
# Removes any visits1 that less than 3,700,000

USAAnalyticsData41720to51620 = USAAnalyticsData41720to51620[USAAnalyticsData41720to51620.visits2 > 3700000]

# Shows all of the records in the dataframe

USAAnalyticsData41720to51620.head(55)
```

```python
# Merges the two datasets together by using an 'outer' join and matches the 'domain' name with each dataframe

result = pd.merge(USAAnalyticsData3620to4420, USAAnalyticsData41720to51620, on=['domain', 'domain'], how='outer')

result.head(55)
```

```python
# Shows the Column, Non-Null Count, and Dtype of each of the fields

result.info()
```

```python
# Fills any result that has a NaN with 0

result = result.fillna(0)

# Shows the first 20 results

result.head(20)
```

```python
# Converts the DatePulled1 and DatePulled2 to datetime

result[["DatePulled1", "DatePulled2"]] = result[["DatePulled1", "DatePulled2"]].apply(pd.to_datetime)

# Shows the Column, Non-Null Count, and Dtype of each of the fields

result.info()
```

```python
# Updates DatePulled1

result['DatePulled1'] = np.where(result['DatePulled1'] == '1970-01-01',pd.to_datetime(['2020-04-05']),result['DatePulled1'])

# Updates DatePulled2

result['DatePulled2'] = np.where(result['DatePulled2'] == '1970-01-01',pd.to_datetime(['2020-05-17']),result['DatePulled2'])

# Shows the results

result
```

```python
# Removes any visits1 that are less than 1.0

result = result[result.visits1 >= 1.0]

result
```

```python
result['VisitsDiff'] = result['visits1'] -  result['visits2']

result['UsersDiff'] = result['users1'] -  result['users2']

result['PageviewsPerSessionDiff'] = result['pageviews_per_session1'] -  result['pageviews_per_session2']

result['AvgSessionDurationDiff'] = result['avg_session_duration1'] -  result['avg_session_duration2']

result['ExitsDiff'] = result['exits1'] -  result['exits2']





result
```

```python
result = result.drop(["DatePulled1", "visits1", "pageviews1", "users1", "exits1", "DatePulled2", "visits2", "pageviews2", "users2", "exits2"], axis=1)

result
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Choose how many bins you want here
num_bins = 20

# Use the histogram function to bin the data
counts, bin_edges = np.histogram(result['avg_session_duration1'], bins=num_bins, normed=True)

# Now find the cdf
cdf = np.cumsum(counts)

# And finally plot the cdf
plt.plot(bin_edges[1:], cdf)

plt.title('CDF (bin=20) for Average Session Duration (Data Pull: 4/5/20)')
plt.xlabel('Average Session Duration (t value)')
plt.ylabel('Probability for t value <= t')
plt.show()
```

```python
# LOG-NORMAL DISTRIBUTION

# standard deviation of normal distribution

sigma = result['pageviews_per_session1'].std()

# mean of normal distribution

mu = result['pageviews_per_session1'].mean()

s = np.random.lognormal(mu, sigma, 75)

count, bins, ignored = plt.hist(s, 75, normed=True, align='mid')

# Calculation

x = np.linspace(min(bins), max(bins), 30)
pdf = (np.exp(-(np.log(x) - mu)**2 / (2 * sigma**2)) / (x * sigma * np.sqrt(2 * np.pi)))

# Plots the info

plt.plot(x, pdf, linewidth=2, color='r')
plt.axis('tight')

plt.title('Log-Normal Distribution of Page Views Per Session (Data Pull: 4/5/20)')
plt.xlabel('Samples')
plt.ylabel('PDF')
plt.show()
```

```python
# Example of the Pearson's Correlation test

from scipy.stats import pearsonr

corrMatrix = result.corr()

data1 = corrMatrix['pageviews_per_session1']
data2 = corrMatrix['pageviews_per_session2']

stat, p = pearsonr(data1, data2)

print('stat=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
    print("\n" + 'Probably independent')
else:
    print("\n" + 'Probably dependent')
```
